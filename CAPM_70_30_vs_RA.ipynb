```python
#monthly return of RA and 70/30 Portfolio

mtl_return = pd.read_csv('mtl_return_70_30_vs_RA.csv', index_col = 0)
mtl_return
```

```python
#quirion vs. 70/30 Performance mit Rebalancing

import matplotlib.pyplot as plt
import statsmodels.api as sm
import datetime as dt

X = mtl_return ['70/30 Performance mit Reb.']
y = mtl_return ['quirion']

X_sm=sm.add_constant(X)
```

```python
model = sm.OLS(y,X_sm)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
import seaborn as sns
```
