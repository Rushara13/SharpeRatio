```python
#Portfoliorendite

dfr = pd.read_csv('Roboadvisor_returns.csv', index_col = 0)
```

```python
# read etf closing prices

dfe = pd.read_csv('MSCI_ACWI_closingPrices.csv', index_col = 0)
```

```python
# calculate returns

returns = dfe.pct_change().dropna()
returns = returns *100
```

```python
#monthly return of RA and MSCI ACWI

mtl_return = pd.read_csv('mtl_return_MSCI_ACWI_RA.csv', index_col = 0)
mtl_return = mtl_return * 100
mtl_return
```

```python
# https://de.investing.com/rates-bonds/usa-government-bonds?maturity_from=40&maturity_to=40
# Stand: 23.Dezember 2021

rfr = 0.0355
```

```python
#calculate excess return 

excess_return = mtl_return - rfr
```

```python
excess_return.mean()
```

```python
#QUIRION SORTINO RATIO BESTIMMEN

excess_return['quirion'].mean()
```

```python
#Rufe quirion spalte auf 

quirion = pd.read_csv('quirion_return.csv', index_col = 0)
quirion
```

```python
quirion ['quirion_neg'] = quirion[quirion ['quirion'] < 0] ['quirion']
quirion ['quirion_neg']
```

```python
quirion ['quirion_neg'].std()
```

```python
sortino_quirion = excess_return['quirion'].mean() / quirion ['quirion_neg'].std()
sortino_quirion
```
