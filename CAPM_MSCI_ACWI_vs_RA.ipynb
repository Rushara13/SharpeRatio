```python
mtl_return = pd.read_csv('mtl_return_MSCI_ACWI_RA.csv', index_col = 0)
mtl_return
```

```python
#quirion vs. MSCI ASWI

import matplotlib.pyplot as plt
import statsmodels.api as sm
import datetime as dt

X = mtl_return ['MSCI ACWI']
y = mtl_return ['quirion']

X_sm=sm.add_constant(X)
```

```python
model = sm.OLS(y,X_sm)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
import seaborn as sns
```

```python
sns.regplot(x='MSCI ACWI', y='quirion', data=mtl_return)
plt.show()
```
