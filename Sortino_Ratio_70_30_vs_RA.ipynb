```python
#monthly return of RA and 70/30 ETF with and without Rebalancing

mtl_return = pd.read_csv('mtl_return_70_30_vs_RA.csv', index_col = 0)
mtl_return = mtl_return * 100
mtl_return
```

```python
# https://de.investing.com/rates-bonds/usa-government-bonds?maturity_from=40&maturity_to=40
# Stand: 23.Dezember 2021

rfr = 0.0355
```

```python
#calculate excess return 

excess_return = mtl_return - rfr
```

```python
excess_return.mean()
```

```python
# Berechnen Sortino Ratio vom 70/30 ETF mit j√§hrliches Rebalancing

#rufe 70/30 mit Reb. spalte auf

mit_Reb = pd.read_csv('70_30_mReb.csv', index_col = 0)
mit_Reb
```

```python
mit_Reb ['mit_Reb_neg'] = mit_Reb[mit_Reb ['70/30 Performance mit Reb.'] < 0] ['70/30 Performance mit Reb.']
mit_Reb['mit_Reb_neg']
```

```python
mit_Reb ['mit_Reb_neg'].std()
```

```python
excess_return['70/30 Performance mit Reb.'].mean()
```

```python
sortino_mit_Reb = excess_return['70/30 Performance mit Reb.'].mean() / mit_Reb ['mit_Reb_neg'].std()
sortino_mit_Reb
```

```python
# Berechnen Sortino Ratio vom 70/30 ETF ohne j√§hrliches Rebalancing

#rufe 70/30 ohne Reb. spalte auf

ohne_Reb = pd.read_csv('70_30_oReb.csv', index_col = 0)
ohne_Reb
```

```python
ohne_Reb ['ohne_Reb_neg'] = ohne_Reb[ohne_Reb ['70/30 Performance ohne Reb.'] < 0] ['70/30 Performance ohne Reb.']
ohne_Reb['ohne_Reb_neg']
```

```python
ohne_Reb ['ohne_Reb_neg'].std()
```

```python
excess_return['70/30 Performance ohne Reb.'].mean()
```

```python
sortino_ohne_Reb = excess_return['70/30 Performance ohne Reb.'].mean() / ohne_Reb ['ohne_Reb_neg'].std()
sortino_ohne_Reb
```
